```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#graphical represantation of arctan() function 
in_array = np.linspace(-np.pi, np.pi, 12)
output_array = np.tan(in_array)
output_array1 = np.arctan(in_array)

print("in_array : ", in_array)
#correct 
print("\n output_array with tan function is", output_array)
print("\n output_array with arctan function is", output_array1)
```

```python
# now red for numpy.arccos()
plt.plot(in_array, output_array, color ="green", marker = "*")
```

```python
plt.plot(in_array, output_array1, color="red", marker="o")
```

```python
# plotting and mixing both two numpy element
plt.title("Blue : numpy.tan() \n red : numpy.arctan()")
```

```python
plt.xlabel("X intercept")
plt.ylabel("Y intercept")
```

```python
plt.show()
```
